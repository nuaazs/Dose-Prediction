```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
root = "/home/zhaosheng/Documents/ScanNetAI/egs/ct2dose/common/niis_selected/RT201561"
img = root + "/ct.nii.gz"
label = root + "/dose.nii.gz"
pred = root + "/struct.nii.gz"

def load_nii_to_array(filename):
    import nibabel as nib
    img = nib.load(filename)
    data = img.get_fdata()
    print(data.shape)
    return data

# plot image
img = load_nii_to_array(img)
# print(img.shape)
img = img[:, 50, :]
plt.figure()
plt.imshow(img, cmap='gray')

pred = load_nii_to_array(pred)
# print(pred.shape)
pred = pred[:, 50, :]
plt.figure()
plt.imshow(pred, cmap='gray')

label = load_nii_to_array(label)
label = label[:, 50, :]
plt.figure()
plt.imshow(label, cmap='gray')
```
